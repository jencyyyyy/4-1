```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../resources/gray_robin.png")
```

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gr = cv2.cvtColor(rgb,cv2.COLOR_BGR2GRAY)
#plt.imshow(img,cmap="gray")
#plt.axis('off')
gr.shape
```

```python
plt.imshow(gr,cmap="gray")
```

```python
#Power law: y=cr^(gamma)
gamma = .3
c=100
power_img  = c*(gr**gamma)


plt.imshow(power_img,cmap="gray")
```

```python
c=100

#log: c*log(1+r)
log_img = c * ((np.log(1 + gr)) / np.log(2))

plt.imshow(log_img,cmap="gray")
```

```python
c=100

#inverse log y: e^(r/c) - 1
inverseLog_img = np.e**(gr/c) - 1

plt.imshow(inverseLog_img,cmap="gray")
```

```python
#use
img_color = cv2.imread('../resources/t_1000.png')
imgResize = cv2.resize(img_color, (512, 512))
img_gray = cv2.cvtColor(imgResize,cv2.COLOR_BGR2GRAY)
```

```python
#use
def smaller(img):
    dsc_image = []
    height,width = img.shape
    for r in range(0, width,2):
        tem_row=[]
        for c in range (0, height, 2):
            pixel = img[r,c]
            tem_row.append(pixel)
        dsc_image.append(tem_row)

    return np.uint8(dsc_image) #uint16, float32 etc
```

```python
#use
img = img_gray.copy()
plt.figure(figsize=(16,7))

for i in range(1,11):
    plt.subplot(2,5,i)
    plt.imshow(img,cmap='gray')
    height,width=img.shape
    plt.title(str(height) + "x" + str(width))
    img = smaller(img)

plt.show()
```

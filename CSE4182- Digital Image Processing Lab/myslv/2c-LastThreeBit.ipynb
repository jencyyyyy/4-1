```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../resources/gray_robin-512.png")

img.shape
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

img.shape
```

```python
img
```

```python
msb3 = (img >>5) <<5
```

the above is just 1 line of bit shifting. 

we can also do masking og 1st 3 bit of each pixel and do the same thing

```python
mask=0b11100000

msbMask = np.bitwise_and(img, mask)
```

```python
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(msb3,cmap="gray")
```

```python
plt.imshow(msbMask,cmap="gray")
```

```python
plt.imshow(cv2.absdiff(img,msb3),cmap="gray")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../resources/gray-apple-square.jpg")
img = cv2.resize(img, (512,512))
```

Why is this?

The answer lies as a caveat with OpenCV.

OpenCV represents RGB images as multi-dimensional NumPy arraysbut in reverse order!

This means that images are actually represented in BGR order rather than RGB!

Theres an easy fix though.

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img
```

```python
print(img.shape)
```

```python
range_start, range_end, add_brightness = 50, 225, 40
height, width = img.shape[0],img.shape[1]
newimg = img.copy()

```

```python
print(newimg.shape)
```

```python
#newimg
```

```python
for r in range(height):
    for c in range(width):
        for channel in range(3):    
            if (newimg[r,c,channel]>=range_start and newimg[r,c,channel]<=range_end):
                if newimg[r,c,channel]>255:
                    newimg[r,c, channel] = 255    
                else:    
                    newimg[r,c, channel]+=add_brightness

            
```

```python
print(newimg.shape)
#newimg
```

```python
plt.imshow(cv2.cvtColor(newimg,cv2.COLOR_RGB2BGR))
```

```python
#plt.figure(8,10)
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.title("original")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(newimg,cv2.COLOR_BGR2RGB))
plt.title("Enhanced")
```

```python
grayImage= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage,cmap='gray')
```

```python
enhancedGray = grayImage.copy()
for r in range(height):
    for c in range(width):    
        if (enhancedGray[r,c]>=range_start and enhancedGray[r,c]<=range_end):
            if enhancedGray[r,c]>255:
                enhancedGray[r,c] = 255    
            else:    
                enhancedGray[r,c]+=add_brightness  

            enhancedGray[c, r] = 255 if enhancedGray[c, r] > 255 else enhancedGray[c, r]          
```

```python
enhancedGray.shape
```

```python
plt.subplot(1,2,1)
plt.imshow(grayImage,cmap='gray')
plt.title("original gray image")
plt.subplot(1,2,2)
plt.imshow(enhancedGray,cmap='gray')
plt.title("Enhanced gray image")
```

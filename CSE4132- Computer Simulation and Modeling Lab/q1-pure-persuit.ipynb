```python
import numpy as np
import matplotlib.pyplot as plt
import pygame
```

```python
def Rand(start, end, num):
	res = []

	for j in range(num):
		res.append(round(np.random.uniform(start, end),2))
        
	return res

# Driver Code test
num = 10
start = 1.0
end = 1000.0
#print(Rand(start, end, num))
bomX =[80, 90, 99, 108, 116, 125, 133, 141, 151, 160, 169, 179, 180]
#bomX = Rand(1,1000,10)

bomY =[0, -2, -5, -9, -15, -18, -23, -29, -28, -25, -21, -20, -17] 
# bomY = Rand(1,1000,10 )
print(bomX)
print(bomY)
fitrX = 0
#fitrX = Rand(1,1000,1)
fitrY = 50
#fitrY = Rand(1,1000,1)

s=20

print(fitrX)
print(fitrY)
```

```python
escape = 900
caught = 10
t=0
xfi = []
yfi = []
while True:

    xfi.append(fitrX)
    yfi.append(fitrY)


    plt.clf()
    plt.title("Simulation of a Pure Pursuit")
    plt.plot(xfi, yfi, marker = "o", label = "Fighter")
    plt.plot(bomX[0 : t+1], bomY[0 : t +1], marker = "o", label = "Bomber")
    plt.xlim(-100, 200)
    plt.ylim(-100, 100)
    plt.legend()
    plt.grid()
    plt.pause(1)

    Dis = ((-bomX[t] + fitrX) ** 2 + (- bomY[t] + fitrY) ** 2) ** 0.5

    print(f"time= {t} xf={round(fitrX,2)} yf= {round(fitrY,2)} xb={round(bomX[t],2)} yb={round(bomY[t],2)} Distance = {Dis}\n")

    if  Dis < caught:
        print(f"caught at {t}, and distance = {Dis}")
        break

    if (t> len(bomX) or (Dis >escape)):
        print(f"escaped on time: {t}")
        break
    
    
    sin = (bomY[t]- fitrY)/Dis
    cos = (bomX[t]- fitrX)/Dis
    t +=1
    fitrX += s*cos
    fitrY += s*sin

```

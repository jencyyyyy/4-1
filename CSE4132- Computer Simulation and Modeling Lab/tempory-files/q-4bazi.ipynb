```python
import matplotlib.pyplot as plt
import numpy as np

# Define the control points
P0 = [0, 0]
P1 = [1, 6]
P2 = [1, -2]
P3 = [3, 3]

# Define the number of points to generate on the curve
n = 100

# Create an array to store the points on the curve
curve = np.zeros((n, 2))

# Generate the points on the curve using the Cubic Bezier curve construction approach
for i in range(n):
    u = i / (n - 1)
    curve[i, 0] = (1 - u)**3 * P0[0] + 3 * u * (1 - u)**2 * P1[0] + 3 * u**2 * (1 - u) * P2[0] + u**3 * P3[0]
    curve[i, 1] = (1 - u)**3 * P0[1] + 3 * u * (1 - u)**2 * P1[1] + 3 * u**2 * (1 - u) * P2[1] + u**3 * P3[1]

# Plot the control points and the curve
plt.plot(P0[0], P0[1], 'ro', label='P0')
plt.plot(P1[0], P1[1], 'go', label='P1')
plt.plot(P2[0], P2[1], 'bo', label='P2')
plt.plot(P3[0], P3[1], 'yo', label='P3')
plt.plot(curve[:, 0], curve[:, 1], 'b-', label='Cubic Bezier Curve')

# Set the axis labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Bezier Curve')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the control points
P0 = np.array([0, 0])
P1 = np.array([2, 2])
P2 = np.array([4, -2])
P3 = np.array([6, 0])

# Define the number of points to evaluate the curve at
n = 100

# Create an array to store the evaluated points
curve = np.zeros((n, 2))

# Evaluate the curve using the Cubic Bezier curve construction approach
for i in range(n):
    t = i / (n - 1)
    B0 = (1 - t)**3
    B1 = 3 * t * (1 - t)**2
    B2 = 3 * t**2 * (1 - t)
    B3 = t**3
    curve[i, :] = B0 * P0 + B1 * P1 + B2 * P2 + B3 * P3

# Plot the curve and the control points
plt.plot(curve[:, 0], curve[:, 1], label='Cubic Bezier Curve')
plt.plot(P0[0], P0[1], 'ro', label='P0')
plt.plot(P1[0], P1[1], 'go', label='P1')
plt.plot(P2[0], P2[1], 'bo', label='P2')
plt.plot(P3[0], P3[1], 'yo', label='P3')
plt.legend()
plt.axis('square')
plt.show()
```
